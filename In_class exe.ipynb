```python
from collections import Counter

# Make sure jobs list exists
companies = []

for job in jobs:
    company = job.get("company")
    if company:   # avoid None or empty values
        companies.append(company)

# Count frequencies safely
company_counts = Counter(companies)

top_3 = company_counts.most_common(3)

print("Q2) Top 3 most frequent companies:")
for company, count in top_3:
    print(company, ":", count)
```

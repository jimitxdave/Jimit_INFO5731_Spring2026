## Q1 (2 points) — Prepare the text data for topic modeling

Using the `clean_text` column, prepare a clean list of documents for BERTopic.

### Steps
- keep only non-null values from `clean_text`
- remove empty strings if there are any
- convert the result to a Python list called `docs`

### Then print
- the number of usable documents
- the first 3 documents

```python
# Write your Answer here (Q1)

# Step 1: Drop null/NaN values from the clean_text column
docs = df["clean_text"].dropna()

# Step 2: Remove empty strings
docs = [doc for doc in docs if doc.strip() != ""]

# Step 3: Print number of usable documents
print("Number of usable documents:", len(docs))

# Step 4: Print first 3 documents
print("\nFirst 3 documents:")
for i, doc in enumerate(docs[:3]):
    print(f"Document {i+1}:", doc)
```

## Q2 (3 points) — Train a BERTopic model

Use the cleaned `docs` list from Q1 and fit a BERTopic model.

### Steps
- create a BERTopic model with `language="english"`
- run `fit_transform(docs)`
- save the outputs in `topics` and `probs`

```python
# Write your Answer here (Q2)

from bertopic import BERTopic

# Step 1: Create BERTopic model with English language
topic_model = BERTopic(language="english")

# Step 2: Fit the model and transform docs
topics, probs = topic_model.fit_transform(docs)

# Step 3: Print confirmation
print("Model training complete!")
print("Total documents processed:", len(topics))
print("Number of topics found:", len(set(topics)))
print("Topics assigned (first 5):", topics[:5])
```

## Q3 (2 points) — Topic summary

Display the output of `get_topic_info()`.

Then report:
1. the total number of **valid topics** discovered (**exclude `-1` if it appears**)
2. the ID of the **largest valid topic** (**exclude `-1`**)

**Hint:** you may want to filter `topic_info` first.

```python
# Write your Answer here (Q3)

# Step 1: Get topic info and display it
topic_info = topic_model.get_topic_info()
print(topic_info)

# Step 2: Filter out the outlier topic (-1)
valid_topics = topic_info[topic_info["Topic"] != -1]

# Step 3: Total number of valid topics
num_valid_topics = len(valid_topics)
print(f"\n1. Total number of valid topics (excluding -1): {num_valid_topics}")

# Step 4: ID of the largest valid topic (highest document count)
largest_topic_id = valid_topics.loc[valid_topics["Count"].idxmax(), "Topic"]
print(f"2. ID of the largest valid topic: {largest_topic_id}")
```

## Q4 (2 points) — Inspect one topic

Choose one valid topic ID (**not `-1`**) and print its **top 5 words**.

```python
# Write your Answer here (Q4)

# Step 1: Choose one valid topic ID (using largest topic from Q3)
chosen_topic_id = largest_topic_id

# Step 2: Get top 5 words for that topic
top_words = topic_model.get_topic(chosen_topic_id)[:5]

# Step 3: Print the top 5 words
print(f"Top 5 words for Topic {chosen_topic_id}:")
print("-" * 35)
for i, (word, score) in enumerate(top_words):
    print(f"  {i+1}. {word}  (score: {score:.4f})")
```

Create a new DataFrame with these two columns:

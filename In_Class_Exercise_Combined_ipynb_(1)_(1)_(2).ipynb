### Q3. Create a sentiment label from the compound score (3 points)
Create a new column called `PredictedSentiment` using this rule:
- `compound >= 0.05` → `positive`
- `compound <= -0.05` → `negative`

```python
from google.colab import files
import pandas as pd

# Step 1: Upload file
uploaded = files.upload()

# Step 2: Load file
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Step 3: Check columns
print("Columns in dataset:", df.columns)

# Step 4: Find correct compound column automatically
compound_col = None
for col in df.columns:
    if col.lower() == "compound":
        compound_col = col

# If not found, stop with message
if compound_col is None:
    print("No 'compound' column found in dataset!")
else:
    # Step 5: Create sentiment label
    def get_sentiment(x):
        if x >= 0.05:
            return "positive"
        elif x <= -0.05:
            return "negative"
        else:
            return "neutral"

    df["PredictedSentiment"] = df[compound_col].apply(get_sentiment)

    # Step 6: Show result
    display(df[["Text", compound_col, "PredictedSentiment"]].head(10))
```

### Q4. Compare review score and sentiment (2 points)
Group by `Score` and compute the **average compound score** for each score value.

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon (only once)
nltk.download('vader_lexicon')

# Load dataset
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Create compound score from Text
df["compound"] = df["Text"].apply(lambda x: sia.polarity_scores(str(x))["compound"])

# Group by Score and compute average compound
result = df.groupby("Score")["compound"].mean().reset_index()

# Show result
result
```

## Reflection (optional, no extra points)
Briefly explain the difference between:
- sentiment analysis
- text classification
- prompt engineering

Sentiment analysis is the process of identifying the emotional tone in text, such as positive, negative, or neutral. It focuses specifically on opinions and feelings expressed in the content.

Text classification is a broader task where any text is assigned to predefined categories, such as spam vs not spam, topic labels, or sentiment. Sentiment analysis is actually a type of text classification.

Prompt engineering is the practice of designing effective instructions (prompts) for AI models to get better and more accurate outputs. It focuses on how you ask the question rather than analyzing the text itself.
